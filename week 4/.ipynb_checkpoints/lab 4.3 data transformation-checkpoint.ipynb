```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, OrdinalEncoder
```

```python
df= pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
```

```python
df
```

```python
numerical = df.select_dtypes(include=np.number)
categorical = df.select_dtypes(include=object)

corr_df = numerical.corr()
sns.heatmap(corr_df,annot=True,mask=np.triu(corr_df))
```

```python
def ord_encoding(x):
    dict_encoding = {
        'jan': 1,
        'feb': 2,
        'small': 1,
        'medsize': 2,
        'large': 3,
        'high school or below': 1,
        'college': 2,
        'bachelor': 3,
        'master': 4,
        'doctor': 5,
    }
    x = x.apply(lambda y: dict_encoding[y])
    return x

ord_encoded = categorical[['education','month','vehicle_size']]
ord_encoded = ord_encoded.apply(ord_encoding)
ord_encoded
```

```python
categorical
```

```python
hot_encoded = categorical.drop(['education','month','vehicle_size'], axis=1)
hot_encoded = pd.get_dummies(hot_encoded, drop_first=True)
hot_encoded.info()
```

```python
cat_encoded = pd.concat([ord_encoded,hot_encoded], axis=1)
cat_encoded
```
