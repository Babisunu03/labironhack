```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
my_playlist =get_playlist_tracks('spotify','5NiEi4iYkHHAwvtmR3AdUe')
```

```python
my_playlist_id_list = []
for i in my_playlist:
    my_playlist_id_list.append(i['track']['id'])
```

```python
audio_list =[]
for i in my_playlist_id_list:
    audio_list.append(sp.audio_features(i)[0])
audio_list
```

```python
import pandas as pd
```

```python
my_playlist_df = pd.DataFrame(audio_list)
my_playlist_df
```
