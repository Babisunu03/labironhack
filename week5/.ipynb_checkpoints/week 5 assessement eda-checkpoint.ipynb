```python

property_data = [
    {'Type': 'house'},
    {'Type': 'unit'},
    {'Type': 'townhouse'},
    
]


property_count = {}
total_properties = len(property_data)

for entry in property_data:
    property_type = entry['property_type']
    if property_type in property_count:
        property_count[property_type] += 1
    else:
        property_count[property_type] = 1


most_common_property = max(property_count, key=property_count.get)
most_common_count = property_count[most_common_property]


percentage_representation = (most_common_count / total_properties) * 100

print(f"The most common property type is '{most_common_property}'")
print(f"It represents {percentage_representation:.2f}% of the properties.")
```

```python
import numpy as np
from scipy import stats

# Sample data (replace with your actual data)
num_rooms = [1, 2, 3, 4, 5]
prices = [50000, 75000, 100000, 120000, 150000]

# Calculate the Spearman's correlation coefficient
correlation_coefficient, p_value = stats.spearmanr(num_rooms, prices)

# Print the correlation coefficient and p-value
print(f"Spearman's correlation coefficient: {correlation_coefficient:.2f}")
print(f"P-value: {p_value:.2f}")

# Interpret the result
if p_value < 0.05:
    if correlation_coefficient > 0:
        print("There is a significant positive correlation.")
    elif correlation_coefficient < 0:
        print("There is a significant negative correlation.")
    else:
        print("There is no significant correlation.")
else:
    print("The correlation is not significant.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr  # For Pearson correlation coefficient

# Sample data (replace with your actual data)
distances_from_cbd = [1.2, 2.5, 3.0, 0.8, 5.0, 2.2, 4.5, 1.0, 3.5]
property_prices = [250000, 320000, 290000, 220000, 410000, 330000, 380000, 230000, 300000]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(distances_from_cbd, property_prices, color='b', alpha=0.7)
plt.xlabel('Distance from CBD (miles)')
plt.ylabel('Property Price ($)')
plt.title('Property Price vs. Distance from CBD')

# Calculate the Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(distances_from_cbd, property_prices)

# Display the correlation coefficient on the plot
plt.text(1.5, 350000, f'Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=12)

# Show the scatter plot
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the correlation coefficient and p-value
print(f"Pearson's correlation coefficient: {correlation_coefficient:.2f}")
print(f"P-value: {p_value:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr  # For Pearson correlation coefficient

# Sample data (replace with your actual data)
property_sizes = [1200, 1500, 1800, 1000, 2200, 1600, 2000, 1100, 1900]
property_prices = [250000, 320000, 290000, 220000, 410000, 330000, 380000, 230000, 300000]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(property_sizes, property_prices, color='b', alpha=0.7)
plt.xlabel('Property Size (sq. ft)')
plt.ylabel('Property Price ($)')
plt.title('Property Price vs. Property Size')

# Calculate the Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(property_sizes, property_prices)

# Display the correlation coefficient on the plot
plt.text(1400, 350000, f'Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=12)

# Show the scatter plot
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the correlation coefficient and p-value
print(f"Pearson's correlation coefficient: {correlation_coefficient:.2f}")
print(f"P-value: {p_value:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import statsmodels.api as sm

# Sample property price data (replace with your actual data)
property_prices = [250000, 320000, 290000, 220000, 410000, 330000, 380000, 230000, 300000, 280000]

# Create a Q-Q plot to compare the distribution to a normal distribution
plt.figure(figsize=(8, 6))
sm.qqplot(np.array(property_prices), line='s', ax=plt.gca())
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.title('Q-Q Plot of Property Prices')

# Calculate skewness and kurtosis
skewness = stats.skew(property_prices)
kurtosis = stats.kurtosis(property_prices)

plt.grid(True)
plt.tight_layout()
plt.show()

# Print skewness and kurtosis
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python
import pandas as pd

# Load your dataset (replace 'your_dataset.csv' with your actual file)
df = pd.read_csv('raw.githubusercontent.com_dipalira_Melbourne-Housing-Data-Kaggle_master_Data_Melbourne_housing_FULL.csv')

# Assuming your dataset has columns named 'Region' and 'Price'
# You can modify these column names to match your dataset
region_column = 'Regionname'
price_column = 'Price'

# Find the region with the highest price
highest_price_region = df[df[price_column] == df[price_column].max()][region_column].values[0]

# Print the result
print(f"The region with the highest price is {highest_price_region} with a price of {df[price_column].max()}")
```

```python
lowest_price_region = df[df[price_column] == df[price_column].min()][region_column].values[0]
print(f"The region with the lowest price is {lowest_price_region} with a price of {df[price_column].min()}")
```

```python
import pandas as pd

# Load your dataset (replace 'your_dataset.csv' with your actual file)


# Assuming your dataset has columns named 'Date' and 'Houses_Sold'
# You can modify these column names to match your dataset
date_column = 'Date'
houses_sold_column = 'SellerG'

# Convert the 'Date' column to a datetime object
df[date_column] = pd.to_datetime(df[date_column])

# Extract the month from each date and create a new column
df['Month'] = df[date_column].dt.month

# Group the data by month and calculate the total number of houses sold for each month
monthly_sales = df.groupby('Month')[houses_sold_column].sum()

# Find the month with the highest number of houses sold
max_month = monthly_sales.idxmax()
# Print the result
print(f"The month with the most houses sold is month {max_month}.")
```
